```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud
```

# Boston

```python
#Importing Dataset

df = pd.read_csv("Boston-top_amenities.csv")

#Checking the Data

df.head()

#Creating the text variable
data = dict(zip(df['amenities'].tolist(), df['appearances'].tolist()))

# Creating word_cloud with text as argument in .generate() method
wc = WordCloud(collocations=False, background_color='white', width=800, height=400).generate_from_frequencies(data)

#word_cloud2 = WordCloud(collocations = False, background_color = 'white').generate(text2)

# Display the generated Word Cloud

plt.figure(figsize=(20,20))
plt.imshow(wc, interpolation='bilinear')

plt.axis("off")
plt.savefig('Boston_wordcloud.png', bbox_inches='tight')
plt.show()
```
